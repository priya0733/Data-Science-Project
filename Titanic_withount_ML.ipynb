```python
import pandas as pd
```

```python
data=pd.read_csv("/content/Titanic-Dataset.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
#replace null values with mean and mode
data['Age'].fillna(data['Age'].mean(),inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0],inplace=True)
```

```python
data
```

```python
#descriptive statistics
data.describe()
```

```python
#frequency count of categorical variables
data['Embarked'].value_counts()
```

```python
data['Sex'].value_counts()
```

```python
data['Pclass'].value_counts()
```

```python
#EDA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#survival count
sns.countplot(x='Survived',data=data)
plt.title('Survived Count')
plt.show()
```

```python
#age distribution
sns.histplot(x='Age',data=data,kde=True)
plt.title('Age Distribution')
plt.show()
#kde represents data using continuous probability density function / curve in one or more dimensions
#kde is generally used for visualisation of observation in a dataset.
```

```python
#fare distribution
sns.histplot(x='Fare',data=data,kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
# Survival rate by gender
sns.barplot(x='Sex', y='Survived', data=data)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
#survival rate by Pclass
sns.barplot(x='Pclass', y='Survived', data=data)
plt.title('Survival Rate by Pclass')
plt.show()
```

```python
#correlation matrix
#correlation matrix
# Select only numerical features for correlation analysis
numerical_data = data.select_dtypes(include=['number'])

corr_matrix = numerical_data.corr()
sns.heatmap(corr_matrix, annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#survival rate by Pclass and Sex
survival_by_class_sex = data.pivot_table('Survived',['Pclass','Sex'],aggfunc='mean')
print(survival_by_class_sex)
```

```python
#survival rate by age group
data['Age_group']=pd.cut(data['Age'],bins=[0,12,18,30,50,80],labels=['child','Teen','yound_adult','adult','senior'])
survival_by_age=data.pivot_table('Survived','Age_group',aggfunc='mean')
print(survival_by_age)
```

```python
# Stacked bar plot for survival by class and gender
pd.crosstab([data['Pclass'],data['Sex']],data['Survived'],margins=True).plot(kind='bar',stacked=True)
plt.title('Survival by Class and Gender')
plt.xlabel('Pclass')
plt.ylabel('count')
plt.show()
```

```python
# Pairplot to explore relations between multiple numerical features
sns.pairplot(data[['Age','Fare']])
plt.show()
```
